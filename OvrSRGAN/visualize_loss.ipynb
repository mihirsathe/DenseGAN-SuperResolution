```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loss_log_file = './loss_log.npy'
loss_log = np.load(loss_log_file)
cols = ['epoch','gen_disc_mse','gen_celoss','gen_mse','disc_mse','disc_acc']
loss_df = pd.DataFrame(loss_log,columns=cols)

plt.figure
plt.subplot(2,1,1)
plt.scatter(loss_df.epoch, loss_df.disc_mse)
plt.scatter(loss_df.epoch, loss_df.gen_celoss)
plt.title('Training Loss - SRGAN')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Discriminator MSE','Generator Cross Entropy'])

plt.subplot(2,1,2)
plt.scatter(loss_df.epoch, loss_df.disc_mse)
plt.scatter(loss_df.epoch, loss_df.gen_celoss)
plt.title('Training Loss - SRGAN')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Discriminator MSE','Generator Cross Entropy'])
plt.show
```
