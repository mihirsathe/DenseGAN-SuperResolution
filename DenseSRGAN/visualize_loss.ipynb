```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
dir_pfx = './'
```

```python
data_dir = dir_pfx + 'loss_logging/OH/'
loss_log_file_o = data_dir + 'loss_log.npy'
loss_log_o = np.load(loss_log_file_o)

cols = ['epoch','batch_idx','loss_hr','acc_hr','loss_gen','acc_gen','loss_adv','acc_adv']
loss_df_o = pd.DataFrame(loss_log_o,columns=cols)

plt.figure
plt.scatter(range(len(loss_df_o.acc_hr)), loss_df_o.loss_hr)
plt.scatter(range(len(loss_df_o.acc_gen)), loss_df_o.loss_gen)
plt.title('Training Accuracy - SRGAN')
plt.xlabel('Batch')
plt.ylabel('Accuracy')
plt.legend(['Disc Loss Real','Disc Loss Generated'])
plt.show
```

```python
loss_df_oe = loss_df_o.drop_duplicates('epoch', keep='last')

plt.figure
plt.scatter(loss_df_oe.epoch, loss_df_oe.loss_hr)
plt.scatter(loss_df_oe.epoch, loss_df_oe.loss_gen)
plt.title('Training Accuracy - SRGAN')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Disc Loss Real','Disc Loss Generated'])
plt.show
plt.show
```
