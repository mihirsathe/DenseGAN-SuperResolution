```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#from google.colab import drive
#drive.mount('/content/drive')

#dir_pfx = '/content/drive/'
#data_dir = '../data/Vehicules1024/'

#dir_pfx = '/content/drive/Team Drives/ECE285_Project/code/prototype/DenseSRGAN/'
#data_dir = dir_pfx + '../data/Vehicules1024/'

dir_pfx = './'
```

```python
data_dir = dir_pfx + 'loss_logging/VEH/'
loss_log_file_v = data_dir + 'loss_log.npy'
loss_log_v = np.load(loss_log_file_v)

cols = ['epoch','batch_idx','loss_hr','acc_hr','loss_gen','acc_gen','loss_adv','acc_adv']
loss_df_v = pd.DataFrame(loss_log_v,columns=cols)

plt.figure
plt.scatter(range(len(loss_df_v.acc_hr)), loss_df_v.loss_hr)
plt.scatter(range(len(loss_df_v.acc_gen)), loss_df_v.loss_gen)
plt.title('Training Accuracy - SRGAN')
plt.xlabel('Batch')
plt.ylabel('Accuracy')
plt.legend(['Disc Loss Real','Disc Loss Generated'])
plt.show
```

```python
loss_df_ve = loss_df_v.drop_duplicates('epoch', keep='last')

plt.figure
plt.scatter(loss_df_ve.epoch, loss_df_ve.loss_hr)
plt.scatter(loss_df_ve.epoch, loss_df_ve.loss_gen)
plt.title('Training Accuracy - SRGAN')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Disc Loss Real','Disc Loss Generated'])
plt.show
#plt.legend(['Adv Loss','Disc Loss on Generated'])
plt.show
```
