```
loss_log_file = dir_pfx + 'loss_log.npy'
loss_log = np.load(loss_log_file)
cols = ['epoch','batch_idx','loss_hr','acc_hr','loss_gen','acc_gen','loss_adv','acc_adv']
loss_df = pd.DataFrame(loss_log,columns=cols)

plt.figure
plt.scatter(range(len(loss_df.acc_hr)), loss_df.acc_hr)
plt.scatter(range(len(loss_df.acc_gen)), loss_df.acc_gen)
plt.title('Training Accuracy - SRGAN')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Accuracy Orig','Accuracy Generated'])
plt.show
```
